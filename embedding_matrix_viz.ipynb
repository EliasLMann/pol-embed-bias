```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import umap
import pickle
```

```python
# load embedding matrices
numberbatch_pickle_file = "embeddings/numberbatch_matrix.pickle"
glove_pickle_file = "embeddings/glove_matrix.pickle"

with open(numberbatch_pickle_file, 'rb') as f:
    # load the pickled embedding matrix
    numberbatch_embed_matrix = pickle.load(f)
    # print a message to confirm the loading
    print(f"Embedding matrix loaded from {numberbatch_pickle_file}")

# open the pickle file for reading in binary mode
with open(glove_pickle_file, 'rb') as f:
    # load the pickled embedding matrix
    glove_embed_matrix = pickle.load(f)
    # print a message to confirm the loading
    print(f"Embedding matrix loaded from {glove_pickle_file}")

print(f"Numberbatch embedding matrix shape: {numberbatch_embed_matrix.shape}")
print(f"GloVe embedding matrix shape: {glove_embed_matrix.shape}")
```

```python
def plot_embedding_matrix(embedding_matrix, n_components=2, labels=None, title=None):

    umap_reducer = umap.UMAP(n_components=n_components)

    # Fit and transform the embedding matrix using UMAP
    umap_embeddings = umap_reducer.fit_transform(embedding_matrix)

    # Plot the UMAP embeddings using Matplotlib
    plt.scatter(umap_embeddings[:, 0], umap_embeddings[:, 1], s=5)

    # Add labels to the plot if provided
    if labels is not None:
        for i, label in enumerate(labels):
            plt.annotate(label, (umap_embeddings[i, 0], umap_embeddings[i, 1]), fontsize=8)

    if title: plt.title(title)
    plt.xlabel('UMAP Component 1')
    plt.ylabel('UMAP Component 2')
    plt.show()
```

```python
plot_embedding_matrix(numberbatch_embed_matrix, title='Numberbatch Embedding Matrix')
```
