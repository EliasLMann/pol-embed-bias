```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data/bias_scores.csv')

# find and print values for Asociated Press url
# print(df[df['domain'] == 'apnews.com'])

# drop [budak_score, fb_score, mturk_score]
# df = df.drop(columns=['budak_score', 'fb_score', 'mturk_score'])

# drop null values
# df = df.dropna()

print(f"Number of rows: {df.shape[0]}")

# print summary statistics
df.describe()
```

```python
# get just the bias scores for New York Times, Breitbart, CNN, Business Insider, the Atlantic, Fox News, Talking Points Memo, Buzzfeed News, National Review, New York Post, the Guardian, NPR, Reuters, Vox, and the Washington Post
df = df[df['domain'].isin(['nytimes.com', 'breitbart.com', 'cnn.com', 'businessinsider.com', 'theatlantic.com', 'foxnews.com', 'talkingpointsmemo.com', 'buzzfeednews.com', 'nationalreview.com', 'nypost.com', 'theguardian.com', 'npr.org', 'reuters.com', 'vox.com', 'washingtonpost.com'])]

# sort by score
df = df.sort_values(by=['score'])

df
```

```python
#drop all rows except vox.com, foxnews.com, and businessinsider.com
new_df = df[df['domain'].isin(['vox.com', 'foxnews.com', 'businessinsider.com'])]
```

```python
# drop rounded_n_shares	rounded_n_accts	fb_score	mturk_score	budak_score	allsides_score	allsides_score_community	pew_score
new_df = new_df.drop(columns=['rounded_n_shares', 'rounded_n_accts', 'fb_score', 'mturk_score', 'budak_score', 'allsides_score', 'allsides_score_community', 'pew_score'])
```
